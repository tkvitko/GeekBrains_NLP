```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_pickle('../hw_1/dummy.pkl')
```

```python
df.head()
```

```python
def list_to_string(source_list):
    return ' '.join(source_list)
```

```python
df['tweet_stemmed_string'] = df['tweet_stemmed'].apply(list_to_string)
```

```python
df['tweet_lemmatized_string'] = df['tweet_lemmatized'].apply(list_to_string)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
count_vectorizer = CountVectorizer(ngram_range=[1,1], 
                                   analyzer='word', 
                                   binary=False, 
                                   tokenizer=str.split, 
                                   max_df=0.9,
                                   max_features = 1000,
                                   stop_words='english')
```

```python
# Создаем the Bag-of-Words модель
bag_of_words = count_vectorizer.fit_transform(df['tweet_stemmed_string'])

# Отобразим Bag-of-Words модель как DataFrame
feature_names = count_vectorizer.get_feature_names()
pd.DataFrame(bag_of_words.toarray(), columns = feature_names)
```

```python
# Создаем the Bag-of-Words модель
bag_of_words = count_vectorizer.fit_transform(df['tweet_lemmatized_string'])

# Отобразим Bag-of-Words модель как DataFrame
feature_names = count_vectorizer.get_feature_names()
pd.DataFrame(bag_of_words.toarray(), columns = feature_names)
```

```python
tfidf_vectorizer = TfidfVectorizer(ngram_range=[1,1], 
                                   analyzer='word', 
                                   binary=False, 
                                   tokenizer=str.split, 
                                   max_df=0.9,
                                   max_features = 2000,
                                   stop_words='english')
```

```python
# Создаем the Bag-of-Words модель
bag_of_words = tfidf_vectorizer.fit_transform(df['tweet_stemmed_string'])

# Отобразим Bag-of-Words модель как DataFrame
feature_names = tfidf_vectorizer.get_feature_names()
pd.DataFrame(bag_of_words.toarray(), columns = feature_names)
```

```python
# Создаем the Bag-of-Words модель
bag_of_words = tfidf_vectorizer.fit_transform(df['tweet_lemmatized_string'])

# Отобразим Bag-of-Words модель как DataFrame
feature_names = tfidf_vectorizer.get_feature_names()
pd.DataFrame(bag_of_words.toarray(), columns = feature_names)
```

```python
# Загружаем данные
data = open('corpus').read()
labels, texts = [], []
for i, line in enumerate(data.split("\n")):
    content = line.split()
    labels.append(content[0])
    texts.append(" ".join(content[1:]))

# создаем df
trainDF = pd.DataFrame()
trainDF['text'] = texts
trainDF['label'] = labels
trainDF.head(5)
```

```python
from sklearn import model_selection, preprocessing, linear_model
from sklearn.metrics import accuracy_score

train_x, valid_x, train_y, valid_y = model_selection.train_test_split(trainDF['text'], trainDF['label'])

# labelEncode целевую переменную
encoder = preprocessing.LabelEncoder()
train_y = encoder.fit_transform(train_y)
valid_y = encoder.fit_transform(valid_y)
```

```python
count_vectorizer.fit(trainDF['text'])

xtrain_count =  count_vectorizer.transform(train_x)
xvalid_count =  count_vectorizer.transform(valid_x)

classifier = linear_model.LogisticRegression()
classifier.fit(xtrain_count, train_y)
predictions = classifier.predict(xvalid_count)
# predictions
accuracy_score(valid_y, predictions)
```

```python
tfidf_vectorizer.fit(trainDF['text'])

xtrain_count =  tfidf_vectorizer.transform(train_x)
xvalid_count =  tfidf_vectorizer.transform(valid_x)

classifier_tfidf = linear_model.LogisticRegression()
classifier_tfidf.fit(xtrain_count, train_y)
predictions_tfidf = classifier_tfidf.predict(xvalid_count)
# predictions
accuracy_score(valid_y, predictions_tfidf)
```

```python
df_dict = {'vectorizer': ['count_vectorizer', 
                          'tfidf_vectorizer'],
           'accuracy': [accuracy_score(valid_y, predictions), 
                        accuracy_score(valid_y, predictions_tfidf)]
          }

result_df = pd.DataFrame.from_dict(df_dict)
result_df
# Увличение количества фич при вектаризации положительно сказывается на accuracy.
```
